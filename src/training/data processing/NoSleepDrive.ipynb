```python
import os
import hashlib

ALLOWED_EXTS = (".jpg", ".jpeg", ".png")

def file_hash(path, chunk_size=8192):
    hasher = hashlib.sha1()
    with open(path, "rb") as f:
        while True:
            chunk = f.read(chunk_size)
            if not chunk:
                break
            hasher.update(chunk)
    return hasher.hexdigest()

def collect_hashes(split_dir):
    hashes = set()
    for root, _, files in os.walk(split_dir):
        for fname in files:
            if fname.lower().endswith(ALLOWED_EXTS):
                full_path = os.path.join(root, fname)
                hashes.add(file_hash(full_path))
    return hashes


# Define split directories
TRAIN_DIR = os.path.join(MRL_DIR, "train")
TEST_DIR  = os.path.join(MRL_DIR, "test")
VAL_DIR   = os.path.join(MRL_DIR, "validate")

# Collect hashes
train_hashes = collect_hashes(TRAIN_DIR)
test_hashes  = collect_hashes(TEST_DIR)
val_hashes   = collect_hashes(VAL_DIR)

# Check overlaps
train_test_overlap = train_hashes & test_hashes
train_val_overlap  = train_hashes & val_hashes
test_val_overlap   = test_hashes & val_hashes

print("Train ∩ Test overlap:", len(train_test_overlap))
print("Train ∩ Val overlap:", len(train_val_overlap))
print("Test ∩ Val overlap:", len(test_val_overlap))

if train_test_overlap or train_val_overlap or test_val_overlap:
    print("❌ OVERLAP DETECTED — DATASET IS CONTAMINATED")
else:
    print("✅ NO IMAGE-LEVEL OVERLAP — DATASET IS CLEAN")
```

```python
def remove_duplicates(split_dir, duplicate_hashes):
    removed = 0
    for root, _, files in os.walk(split_dir):
        for fname in files:
            if fname.lower().endswith(ALLOWED_EXTS):
                path = os.path.join(root, fname)
                if file_hash(path) in duplicate_hashes:
                    os.remove(path)
                    removed += 1
    return removed

removed_from_test = remove_duplicates(TEST_DIR, train_hashes)
removed_from_val  = remove_duplicates(VAL_DIR, train_hashes)

print("Removed from test:", removed_from_test)
print("Removed from val:", removed_from_val)
```

```python
from PIL import Image
import os
import random

def check_random_samples(root, n=5):
    for split in ["train", "validate", "test"]:
        split_dir = os.path.join(root, split)

        class_dirs = [
            d for d in os.listdir(split_dir)
            if os.path.isdir(os.path.join(split_dir, d))
        ]

        print(f"\nChecking {split} split")

        for cls in class_dirs:
            cls_dir = os.path.join(split_dir, cls)
            images = [
                f for f in os.listdir(cls_dir)
                if f.lower().endswith((".jpg", ".jpeg", ".png"))
            ]

            if len(images) == 0:
                print(f"⚠️ No images in {cls_dir}")
                continue

            for img_name in random.sample(images, min(n, len(images))):
                img_path = os.path.join(cls_dir, img_name)
                img = Image.open(img_path)

                assert img.size == (64, 64), f"Wrong size: {img.size}"
                assert img.mode == "RGB", f"Wrong mode: {img.mode}"

    print("\n✅ Sanity check passed — all images are 64×64 RGB")

check_random_samples(MRL_DIR)
```

```python
import os

def count_images(split_dir):
    count = 0
    for root, _, files in os.walk(split_dir):
        for f in files:
            if f.lower().endswith((".jpg", ".jpeg", ".png")):
                count += 1
    return count

for split in ["train", "validate", "test"]:
    split_path = os.path.join(MRL_DIR, split)
    print(split, count_images(split_path))
```
